```python
import numpy as np
out = []
for data in alldata['gp']:
    results = data['results']
    out.append([data['ls'],data['accuracy'],np.diff(data['abCI'])[0],np.max([np.max(res[0]) for res in results])])
out = np.array(out)
out = out[np.argsort(out[:,0]),:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(out[:,0],out[:,2],'-k',lw=2)
plt.plot(out[:,0],out[:,3])
plt.figure()
plt.plot(out[:,0],out[:,1])
```
